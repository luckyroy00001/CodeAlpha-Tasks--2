# Training And Testing

```python
from sklearn.model_selection import train_test_split
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Norm_MMS = MinMaxScaler()
```

```python
Norm_MMS_train = Norm_MMS.fit_transform(X_train)
```

```python
Norm_MMS_tset = Norm_MMS.fit_transform(X_test)
```

```python
Norm_MMS_train_dataFrame = pd.DataFrame(Norm_MMS_train,columns=['Pregnancies','Glucose','BloodPressure'])
```
